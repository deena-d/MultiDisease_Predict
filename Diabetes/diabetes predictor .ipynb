Importing the libraries


```python
import numpy as np
import pandas as pd
import seaborn as sb
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

Data collection and Analysis

```python
diabetes_data = pd.read_csv("/content/diabetes.csv")
```

```python
diabetes_data.head()
```

```python
diabetes_data.shape
```

1 --- diabetes

0 --- healthy

```python
diabetes_data.isnull()
```

```python
diabetes_data['Outcome'].value_counts()
```

```python
diabetes_data.groupby('Outcome').mean()
```

```python
x = diabetes_data.drop(['Outcome'],axis= 1)
y = diabetes_data['Outcome']
```

```python
print(x)
```

```python
print(y)
```

Data Standardization

```python
Scaler = StandardScaler()
```

```python
standardized  = Scaler.fit_transform(x)
```

```python
print(standardized)
```

```python
x = standardized
```

training and test model

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, stratify=y, random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Training the  Model

```python
classifier  = svm.SVC(kernel='linear')
```

```python
classifier.fit(x_train,y_train)
#trained model stored in the classifier
```

Model Evaluation

```python
x_train_prediction = classifier.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy score ',training_data_accuracy)
```

```python
x_test_prediction = classifier.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('accuracy score testing ',test_data_accuracy)
```

```python
input_data= (9,170,74,31,0,44,0.403,43)

input_data_as_numpy_array = np.asarray(input_data)

input_data_reshaped= input_data_as_numpy_array.reshape(1,-1)

std_data = Scaler.transform(input_data_reshaped)

prediction = classifier.predict(std_data)

print(prediction)

if (prediction [0]== 0):
  print("A Person is healthy")
else:
  print("That person have the diabetics")
```
