Import dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

data collection and processing

```python
parkinson_data = pd.read_csv('/content/parkinsons.csv')
```

DATA PRE PROCESSING

```python
x = parkinson_data.drop(['name','status'],axis= 1)
y = parkinson_data['status']
```

```python
print(x)
```

```python
print(y)
```

Test and Train data

```python
x_test,x_train,y_test,y_train = train_test_split(x,y,test_size=0.2)
```

```python
print(x.shape,x_test.shape,x_train.shape)
```

Standradization

```python
scalar = StandardScaler()
```

```python
scalar.fit(x_train)
```

```python
x_train= scalar.transform(x_train)

x_test= scalar.transform(x_test)
```

```python
print(x_train)
```

model training

```python
model  = svm.SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```

MODEL EVALUTION

Accuracy Score

```python
x_train_prediction= model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy of the train data:",training_data_accuracy)
```

```python
x_test_prediction= model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy of the test data:",test_data_accuracy)
```

Building the model

```python
input_data= (237.226,247.326,225.227,0.00298,0.00001,0.00169,0.00182,0.00507,0.01752,0.164,0.01035,0.01024,0.01133,0.03104,0.0074,22.736,0.305062,0.654172,-7.31055,0.098648,2.416838,0.095032)

input_data_as_numpy_array = np.asarray(input_data)

input_data_reshaped= input_data_as_numpy_array.reshape(1,-1)

std_data = scalar.transform(input_data_reshaped)

prediction = model.predict(std_data)

print(prediction)


if (prediction [0]== 0):
  print("A Person is healthy")
else:
  print("That person have the Parkinson disease")
```
